```python
import matplotlib.pyplot as plt
import pandas
import matplotlib.legend as lgd
```

```python
# Прочитать данные о продажах за все месяцы и отобразить их в виде линейного графика. 
# Подписать ось икс как “Номер месяца”, игрек - “Прибыль”
```

```python
data = pandas.read_csv("company_sales_data.csv")
data
```

```python
month = data["month_number"].tolist()
month
```

```python
t_profit = data["total_profit"].tolist()
t_profit
```

```python
plt.plot(month, t_profit)
plt.xlabel("Номер месяца")
plt.ylabel("Прибыль")
plt.title("Прибыль по месяцам")
plt.xticks([i for i in range(1,13)])
plt.yticks([i for i in range(150000,350000, 25000)])
plt.show()
```

Прочитать данные о продажах за все месяцы и отобразить их в виде линейного графика. 
Добавить следующие свойства:
- Стиль линии - точечный, цвет линии - красный
- Показать легенду в правом нижнем углу
- ось икс = номер месяца
- ось игрек = продано единиц
- Маркер в виде круга.
- Цвет маркера - зеленый
- Ширина линии = 3
'lower right' = 4


```python
t_units = data["total_units"].tolist()
t_units
```

```python
plt.plot(month, t_units , c = "r", mfc = "g" , marker = "o", linestyle = ":" , linewidth = 3 ,  label='продажи')
plt.xlabel("Номер месяца")
plt.ylabel("продано единиц")
plt.title("Продажи по месяцам")
plt.xticks([i for i in range(1,13)])
plt.yticks([i for i in range(17500,31000, 1000)])
plt.legend(loc = 4)
plt.show()
```

Прочитать данные о продажах зубной пасты за каждый месяц и отобразить в виде точечной диаграммы. 
Добавить сетку со стилем “--”.  Добавить легенду, подписи осей и заголовок диаграммы.

```python
toothpaste = data["toothpaste"].tolist()
toothpaste
```

```python
plt.scatter(month, toothpaste , c = "g", marker = "o", linestyle = ":" , linewidth = 3 ,  label='продажи зубной пасты')
plt.xlabel("Номер месяца")
plt.ylabel("продано единиц")
plt.title("Продажи зубной пасты по месяцам")
plt.grid(color='grey', linestyle='--', linewidth=1)
plt.xticks([i for i in range(1,13)])
plt.yticks([i for i in range(4500,8500, 250)])
plt.legend(loc = 0)
plt.show()
```

Прочитать данные о продажах пеня для бритья и средства для мытья лица, показать это в виде столбчатой диаграммы. 
График должен отражать количество проданных товаров на каждый месяц для каждого продукта. 
На один месяц выйдет два столбика. Легенду и подписи, конечно же.  

```python
facecream = data["facecream"].tolist()
facecream 
```

```python
facewash= data["facewash"].tolist()
facewash
```

```python
plt.bar([i-0.25 for i in month], facecream ,width = 0.25 ,  label='крем для бритья', align='edge' )
plt.bar([i+0.25 for i in month], facewash , width = - 0.25 ,  label='средство для лица', align='edge')
plt.xlabel("Номер месяца")
plt.ylabel("продано единиц")
plt.title("Продажи средств для лица по месяцам")
plt.xticks([i for i in range(1,13)])
plt.yticks([i for i in range(0,5000, 500)])
plt.legend(loc = 1)
plt.show()
```
